## Q1 — Statistical consistency

Earlier measurements: $20.1$, $20.2$, $20.5$, $19.8\,\mathrm{g}$. New sample: $20.6\,\mathrm{g}$.  
Test with `is_consistent` using a $2\sigma$ criterion.

```python
data    = [20.1, 20.2, 20.5, 19.8]
new_val = 20.6
result  = pt.is_consistent(new_val, data, n_sigma=2.0)
for k, v in result.items():
    print(f"{k}: {v}")
# consistent: True  ->  Answer: A
```

## Q3 — Two-ball meeting height

Ball 1 is thrown upward from the ground with $v_0 = 50\,\mathrm{m/s}$.  
Ball 2 is dropped simultaneously from $h = 100\,\mathrm{m}$ above Ball 1.

Positions: $y_1(t) = v_0 t - \tfrac{1}{2}gt^2$ and $y_2(t) = 100 - \tfrac{1}{2}gt^2$.  
Setting equal gives $v_0 t = 100 \Rightarrow t = 2\,\mathrm{s}$ (gravity cancels).

```python
v0_ball1 = 50.0
h_ball2  = 100.0
t_meet   = h_ball2 / v0_ball1
y_meet   = pt.free_fall_y(y0=0.0, v0=v0_ball1, t=t_meet)
print(f"t_meet = {t_meet:.2f} s")
print(f"y_meet = {y_meet:.2f} m")
# Answer: I  (80.4 m)
```

## Q11 — Fusion energy distribution ($D + T \to {}^4\mathrm{He} + n$)

$E_{\mathrm{fusion}} = 17.6\,\mathrm{MeV}$, $m_{\mathrm{He}} = 4m_n$, and initial momenta are approximately zero.  
Momentum conservation: $4m_n v_{\mathrm{He}} = m_n v_n \Rightarrow v_n = 4v_{\mathrm{He}}$.  
Kinetic-energy ratio: $\dfrac{KE_{\mathrm{He}}}{KE_n} = \dfrac{4m_n v_{\mathrm{He}}^2}{m_n(4v_{\mathrm{He}})^2} = \dfrac{1}{4}$, so He gets $1/5 = 20\%$.

```python
result = pt.fusion_energy_split(mass_ratio=4.0, E_total=17.6)
print(f"He (heavy): {result['heavy_fraction'] * 100:.0f}%   {result['E_heavy']:.2f} MeV")
print(f"n  (light): {result['light_fraction'] * 100:.0f}%   {result['E_light']:.2f} MeV")
# Answer: D  (He 20%, n 80%)
```

## Q13 — Work scaling with time

$d = \tfrac{1}{2}at^2 \Rightarrow W = Fd = mad = \dfrac{2md^2}{t^2}$. Thus $W \propto 1/t^2$, so  
$\dfrac{W_2}{W_1} = \dfrac{t_1^2}{t_2^2} = \dfrac{1^2}{3^2} = \dfrac{1}{9}$.

```python
d, m_block = 10.0, 1.0  # arbitrary; only the ratio matters
W1 = pt.constant_F_to_travel_d_in_t(d=d, t=1.0, m=m_block)["W"]
W2 = pt.constant_F_to_travel_d_in_t(d=d, t=3.0, m=m_block)["W"]
print(f"W1 = {W1:.4f} J,  W2 = {W2:.4f} J")
print(f"W2/W1 = {W2/W1:.4f}  (= 1/9 = {1/9:.4f})")
# Answer: B  (W2 = W1/9)
```

## Q14 — Two discs: total mechanical energy vs time

Cannot be solved computationally (qualitative graph question).

**Physics:**
- Phase 1 (disc in free fall, no contact): mechanical energy conserved -> $E_m$ constant.
- Phase 2 (disc lands, inelastic impact): $E_m$ drops discontinuously.
- Phase 3 (kinetic friction between discs, spinning up to common $\omega$): $E_m$ decreases continuously.
- Phase 4 (same $\omega$, no relative slipping): $E_m$ constant again.

Graph: flat -> sharp drop -> gradual decrease -> flat.
