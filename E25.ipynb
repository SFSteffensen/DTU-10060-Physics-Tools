## Q1 — Statistical consistency of measurements

Earlier measurements of identical samples: 20.1 g, 20.2 g, 20.5 g, 19.8 g.  
A new sample measures **20.6 g**. Can we assume it has the same mass?

- A) Yes
- B) Maybe
- C) No

````python
data    = [20.1, 20.2, 20.5, 19.8]
new_val = 20.6
result  = pt.is_consistent(new_val, data, n_sigma=2.0)
for k, v in result.items():
    print(f"{k}: {v}")
````

The sample mean is **20.15 g** with a standard deviation of **≈ 0.289 g**. The new measurement of 20.6 g deviates by ≈ 1.56σ from the mean — well within the 2σ consistency threshold. The value `consistent: True` confirms this.  

**Answer: A — Yes**

## Q3 — Two-ball meeting height

Ball 1 is thrown vertically upward from the ground at $v_0 = 50\ \text{m/s}$.  
Ball 2 is dropped simultaneously from $h = 100\ \text{m}$ above Ball 1.  
At what height do they meet?

- A) 17.0 m
- B) 28.9 m
- C) 36.2 m
- D) 36.9 m
- E) 49.1 m
- F) 52.6 m
- G) 53.3 m
- H) 80.0 m
- I) 80.4 m
- J) 97.8 m

````python
v0_ball1 = 50.0   # m/s upward
h_ball2  = 100.0  # m above ball 1

# y1(t) = v0*t - g*t^2/2,  y2(t) = 100 - g*t^2/2
# Setting equal: v0*t = 100  ->  t = 100/50 = 2 s  (gravity cancels)
t_meet = h_ball2 / v0_ball1
y_meet = pt.free_fall_y(y0=0.0, v0=v0_ball1, t=t_meet)
print(f"Meeting time  = {t_meet:.2f} s")
print(f"Meeting height = {y_meet:.2f} m")
````

Because both balls experience the same gravitational acceleration, the $-\frac{1}{2}g t^2$ terms cancel when equating positions. The meeting condition reduces to $v_0 t = 100\ \text{m}$, giving $t = 2.00\ \text{s}$. Substituting back: $y = 50 \times 2 - \frac{1}{2}(9.82)(4) = 80.36\ \text{m} \approx 80.4\ \text{m}$.  

**Answer: I — 80.4 m**

## Q4 — River boat speed

Boat speed relative to water: $v$. Water current speed: $v_0$.  
Downstream travel of distance $L$: time $T$.  
Upstream return of distance $L$: time $2T$.  
Find $v$.

- A) $v = 4v_0$
- B) $v = 3v_0$
- C) $v = 2v_0$
- D) $v = v_0$
- E) $v = \frac{1}{2}v_0$
- F) $v = \frac{1}{3}v_0$
- G) $v = \frac{1}{4}v_0$

````python
import sympy as sp

v, v0 = sp.symbols("v v0", positive=True)

# Downstream: L = (v + v0) * T
# Upstream:   L = (v - v0) * 2T
# Setting equal and dividing by T:
eq = sp.Eq(v + v0, 2 * (v - v0))
print("Equation:", eq)

sol = sp.solve(eq, v)[0]
print("v =", sol)
````

Setting up $L = (v + v_0)T$ (downstream) and $L = (v - v_0) \cdot 2T$ (upstream), then dividing both sides by $T$ gives $(v + v_0) = 2(v - v_0)$. Expanding and collecting terms: $3v_0 = v$. The symbolic solver confirms $v = 3v_0$.

**Answer: B — $v = 3v_0$**

## Q7 — Two stacked blocks: center-of-mass acceleration

Two blocks, each $m = 1.0\ \text{kg}$, stacked on a frictionless surface.  
$\mu_s = 0.80$, $\mu_k = 0.50$. Horizontal force $F = 20\ \text{N}$ applied to the **top** block.  
What is the acceleration of the common center of mass?

- A) 20 m/s²
- B) 15 m/s²
- C) 12 m/s²
- D) 10 m/s²
- E) 7.0 m/s²
- F) 5.1 m/s²

````python
result = pt.two_blocks_stacked(
    m_top=1.0, m_bot=1.0, F=20.0, mu_s=0.80, mu_k=0.50, pulled="top"
)
print(f"slipping    = {result['slipping']}")
print(f"a_top       = {result['a_top']:.2f} m/s^2")
print(f"a_bot       = {result['a_bot']:.2f} m/s^2")
print(f"a_cm        = {result['a_cm']:.2f} m/s^2")
print(f"friction    = {result['friction']:.3f} N")
print(f"max static  = {0.80 * 1.0 * pt.g:.3f} N")
````

The maximum static friction between the blocks is $\mu_s m g = 0.80 \times 1.0 \times 9.82 \approx 7.86\ \text{N}$. **Answer: D — 10 m/s²**

## Q11 — Fusion energy distribution (He + neutron)

Frictionless floor. Constant force from rest, distance $d$ in time $t$ requires work $W_1$.

## Q14 — Two discs: total mechanical energy vs time

A rotating disc has a stationary disc dropped onto it (no bounce). Friction between the discs brings them to a common angular velocity.  
Which graph describes $E_m(t)$?

*(Multiple graph options A–? shown in the original exam.)*

This is a qualitative question — no computation required.

- **Phase 1** (disc in free fall, no contact): total mechanical energy is conserved — $E_m$ is **flat**.
- **Phase 2** (impact): this is an inelastic collision — mechanical energy is lost instantaneously — $E_m$ **drops discontinuously**.
- **Phase 3** (kinetic friction between the discs as they equalise $\omega$): friction continuously dissipates rotational energy — $E_m$ **decreases gradually**.
- **Phase 4** (same $\omega$, no more slipping): no dissipation — $E_m$ is **flat** again.

The graph showing **flat → sharp drop → gradual decrease → flat** matches option H.  

**Answer: H**
