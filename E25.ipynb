## Q2 — Statistical consistency

Earlier measurements: 20.1, 20.2, 20.5, 19.8 g. New sample: 20.6 g.

```python
data    = [20.1, 20.2, 20.5, 19.8]
new_val = 20.6
result  = pt.is_consistent(new_val, data, n_sigma=2.0)
for k, v in result.items():
    print(f"{k}: {v}")
# consistent: True  ->  Answer: A
```

## Q4 — Two-ball meeting height

Ball 1 thrown upward from ground at v₀ = 50 m/s.  
Ball 2 dropped simultaneously from h = 100 m above Ball 1.

Positions: y₁(t) = v₀ t − g t²/2 and y₂(t) = 100 − g t²/2.

```python
v0_ball1 = 50.0
h_ball2  = 100.0
t_meet   = h_ball2 / v0_ball1
y_meet   = pt.free_fall_y(y0=0.0, v0=v0_ball1, t=t_meet)
print(f"t_meet = {t_meet:.2f} s")
print(f"y_meet = {y_meet:.2f} m")
# Answer: I  (80.4 m)
```

## Q12 — Fusion energy distribution (D + T → ⁴He + n)

E_fusion = 17.6 MeV, m_He = 4 m_n, initial momenta ≈ 0.  
Momentum conservation: 4 m_n v_He = m_n v_n → v_n = 4 v_He.

## Q14 — Work scaling with time

d = ½ a t² → W = F d = m a d = 2 m d² / t². So W ∝ 1/t²,  
meaning W₂/W₁ = t₁²/t₂² = 1²/3² = 1/9.

```python
d, m_block = 10.0, 1.0  # arbitrary; only the ratio matters
W1 = pt.constant_F_to_travel_d_in_t(d=d, t=1.0, m=m_block)["W"]
W2 = pt.constant_F_to_travel_d_in_t(d=d, t=3.0, m=m_block)["W"]
print(f"W1 = {W1:.4f} J,  W2 = {W2:.4f} J")
print(f"W2/W1 = {W2/W1:.4f}  (= 1/9 = {1/9:.4f})")
# Answer: B  (W2 = W1/9)
```
